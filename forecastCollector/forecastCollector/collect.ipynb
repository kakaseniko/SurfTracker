```python
import pandas as pd
from datetime import datetime
```

```python
url = "https://waterberichtgeving.rws.nl/wbviewer/maak_grafiek.php?loc=KUST_28&set=wigo&nummer=7&format=csv"
```

```python
df = pd.read_csv(url, sep = ';', decimal = ',', 
                 names= ['date', 'wavePeriod', 'waveHeight', 'SWAN_swellHeight'], usecols = ['date', 'wavePeriod', 'waveHeight'], parse_dates = ['date'], skiprows = 1)
```

```python
today = datetime.today().strftime('%Y-%m-%d')
df = df.query('date > @today')
```

```python
df = df.drop(df.tail(1).index)
```

```python
df['date'] = pd.to_datetime(df.date, format='%Y-%m-%d %H:%M:%S')
df['wavePeriod'] = df['wavePeriod'].astype(float)
df['wavePeriod'] = df['wavePeriod'] / 10
df['wavePeriod'] = df['wavePeriod'].astype(int)
df['waveHeight'] = df['waveHeight'].astype(float).apply(lambda x: x / 100).round(1)

```

```python
df['date'] = df['date'].dt.strftime('%Y-%m-%d %H:%M:%S')
df.to_json(orient='records')
```
